# Third Task

```python
#for dropping rows which have nan in all cells
df.dropna(how='all')
```

```python
df['Notes']=df['Notes'].replace(to_replace=np.nan,value=0)
df['Status']=df['Status'].replace(to_replace=np.nan,value=0)
```

```python
df['Benefits']=df['TotalPayBenefits']-df['TotalPay']

df['BasePay']=df['TotalPay']-df['OvertimePay']-df['OtherPay']

df['OvertimePay']=df['TotalPay']-df['BasePay']-df['OtherPay']

df['OtherPay']=df['TotalPay']-df['BasePay']-df['OvertimePay']


nan_rows = df.isna().any(axis=1)
temp =[]
for i in range(nan_rows.shape[0]) :
    if nan_rows[i]:
        temp.append(i)

df = df.drop(temp, axis=0)

print(df.isnull().sum())
```

```python
a = df.index[df['TotalPay'] <= 0]
df = df.drop(a, axis=0)
```

```python
df
```

First , I replaced all NAN values with zero in the two columns (Status , Notes) and I notice that the column that contains the most NaN value is the column with the name (benefits), and we notice that the value of column (TotalPayBenefits) is equal to the sum of the values of the columns (benefits) and (TotalPay).
In this way, we can fill in the column (benefits), as we notice that there are no zero values in the previous two columns
Then we fill each column (BasePay), column (OvertimePay), and column (OtherPay) with the value resulting from the following equation:
The sum of the following columns gives the column value (TotalPay)
In addition, I deleted the lines that had more than NAN, and I cannot fill them with the previously deduced equation and I deleted a row which have negative or zero salarie which it is wrong
# why do I use the previus method ?
Through these operations, I did not add any external information about the data set, but rather I deduced it through the relationships between the columns, and in this way the data remained intact and correct.

```python
print(f"Positive correlation between salary and TotalPayBenefits , method = pearson {df['TotalPay'].corr(df['TotalPayBenefits'],method='pearson')}")
print(f"Positive correlation between salary and TotalPayBenefits , method = kendall {df['TotalPay'].corr(df['TotalPayBenefits'],method='kendall')}")
print(f"Positive correlation between salary and TotalPayBenefits , method = spearman {df['TotalPay'].corr(df['TotalPayBenefits'],method='spearman')}")

print(f"Positive correlation between salary and BasePay {df['TotalPay'].corr(df['BasePay'],method='pearson')}")
print(f"Positive correlation between salary and BasePay {df['TotalPay'].corr(df['BasePay'],method='kendall')}")
print(f"Positive correlation between salary and BasePay {df['TotalPay'].corr(df['BasePay'],method='spearman')}")

print(f"correlation between salary and OvertimePay {df['TotalPay'].corr(df['OvertimePay'],method='pearson')}")
print(f"correlation between salary and OvertimePay {df['TotalPay'].corr(df['OvertimePay'],method='kendall')}")
print(f"correlation between salary and OvertimePay {df['TotalPay'].corr(df['OvertimePay'],method='spearman')}")

print(f"correlation between salary and OtherPay {df['TotalPay'].corr(df['OtherPay'],method='pearson')}")
print(f"correlation between salary and OtherPay {df['TotalPay'].corr(df['OtherPay'],method='kendall')}")
print(f"correlation between salary and OtherPay {df['TotalPay'].corr(df['OtherPay'],method='spearman')}")

print(f"correlation between salary and Benefits {df['TotalPay'].corr(df['Benefits'],method='pearson')}")
print(f"correlation between salary and Benefits {df['TotalPay'].corr(df['Benefits'],method='kendall')}")
print(f"correlation between salary and Benefits {df['TotalPay'].corr(df['Benefits'],method='spearman')}")
```

```python
plt.scatter(df['TotalPay'],df['TotalPayBenefits'])
plt.title('correlation between salary and TotalPayBenefits')
plt.show()

plt.scatter(df['TotalPay'],df['BasePay'])
plt.title('correlation between salary and BasePay')
plt.show()

plt.scatter(df['TotalPay'],df['OvertimePay'])
plt.title('correlation between salary and OvertimePay')
plt.show()

plt.scatter(df['TotalPay'],df['OtherPay'])
plt.title('correlation between salary and OtherPay')
plt.show()

plt.scatter(df['TotalPay'],df['Benefits'])
plt.title('correlation between salary and Benefits')
plt.show()
```
